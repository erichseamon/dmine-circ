## 2. Inventory and Harvest Data

In this step, we assess what data might be relevant to the issue at hand.  

<H3>INVENTORY AND HARVEST - Assess problem and assemble universe of data.  </H3>

Our premise:  Insurance claims are filed by farmers nationwide, documenting their monthly losses by commodity (which are summarized and provided to the public at a county level), provides a good estimator of overall food resilience. 

<H4> Assembled Data: </H4>

- response variable: USDA monthly claims of crop loss, by county, by commodity, for the United States, 1980 to present.  this is represented as a set of csv files, one for each year, containing ~ 250,000 records each.
- feature variables: GRIDMET daily calculated variables, averages monthly across each of county of Washington.

<H4> Programming Code: </H4>

 1. R is used to extract GRIDMET climate data, and transform data for machine learning analysis. 
 2. Python is used to run and implement the machine learning models.
